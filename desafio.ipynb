```python
# crindo data frame
vp = pd.read_csv('vendas_pizzaria.csv')
```

```python
# verificando se tem faltantes
vp.isnull().sum()
```

```python
# substituindo o valor de NA pela media
vp["valor_total"].fillna(vp['valor_total'].mean(), inplace=True)
```

```python
# somei a media do valor total
vp["valor_total"].mean()
```

```python
#substituindo o valor de na por zero
vp["valor_total"].fillna(0, inplace=True)
```

```python
# tirando o NA das colunas
vp.dropna(inplace=True)
```

```python
# verificando se ainda existe valores nulos
vp.isnull().sum()
```

```python
# transformando a coluna data em tipo data
vp['data_venda'] = pd.to_datetime(vp['data_venda'])
```

```python
vp['quantidade_pizzas'] = vp['quantidade_pizzas'].clip(lower=0)
```

```python
vp["valor_unitario"] = vp['valor_unitario'].round(2)
```

```python
vp.head(10)
```

```python
# calcular a média sem considerar os zeros
media = vp.loc[vp['valor_unitario'] != 0, 'valor_unitario'].mean()

# substituir 0 pela média
vp['valor_unitario'] = vp['valor_unitario'].replace(0, media).round(2)
```

```python
vp.head(10)
```

```python
vp['id_franquia'] = vp['id_franquia'].astype(int)
```
